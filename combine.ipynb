```python
df1 = pd.read_csv('data/comments/df_1.csv', encoding='utf-8')
df2 = pd.read_csv('data/comments/df_2.csv', encoding='utf-8')
df3 = pd.read_csv('data/comments/df_3.csv', encoding='utf-8')
dfw = pd.read_csv('data/comments/df_w.csv', encoding='utf-8')

df_combined = pd.concat([df1, df2, df3, dfw], ignore_index=True)
print(len(df_combined))
```

```python
df_combined.to_csv('data/comments/combined_0430.csv', index=False, encoding='utf-8')
```

### Clean data

```python
## claude fixed code

def fix_misaligned_comments(df):
    fixed_rows = []
    current_comment = None
    
    for _, row in df.iterrows():
        # Convert all values to strings and handle NaN values
        row_dict = {}
        for col, val in row.items():
            row_dict[col] = str(val) if pd.notna(val) else ''
        
        # Check if this is a new comment (contains the specific pattern)
        if '【#賀瓏夜夜秀】' in row_dict['video_title']:
            # Save previous comment if it exists
            if current_comment is not None:
                fixed_rows.append(current_comment)
            
            # Start a new comment
            current_comment = row_dict.copy()
        else:
            # This is a continuation of the previous comment
            if current_comment is not None:
                # Append the content from video_title to the current comment_text
                if row_dict['video_title'].strip():
                    current_comment['comment_text'] += ' ' + row_dict['video_title'].strip()
                
                # Preserve any other valid data from other columns
                for col in df.columns:
                    if col != 'video_title' and row_dict[col].strip():
                        current_comment[col] = row_dict[col]
    
    # Add the last comment
    if current_comment is not None:
        fixed_rows.append(current_comment)
    
    return pd.DataFrame(fixed_rows)

# Fix the misaligned comments
fixed_df = fix_misaligned_comments(df_combined)
print(len(fixed_df))

# Save the fixed data to a new CSV file
fixed_df.to_csv('data/comments/fixed_comments_combined.csv', index=False, encoding='utf-8-sig')
```
